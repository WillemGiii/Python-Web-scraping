```python
import os
import time
from dotenv import load_dotenv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 載入環境變數
load_dotenv()
ACCOUNT = os.getenv("ACCOUNT")
PASSWORD = os.getenv("PASSWORD")

driver = webdriver.Chrome()

try:
    # 1. 前往 CODIS 首頁/測站頁面
    print("正在前往網站...")
    driver.get("https://codis.cwa.gov.tw/Home?locale=zh-TW")

    # 2. 等待並點擊右上角的「登入」按鈕 (因為登入框通常預設是隱藏的)
    print("等待首頁載入...")
    wait = WebDriverWait(driver, 10) # 設定最長等待 10 秒
    
    # 尋找頁面上的登入按鈕 (這裡假設按鈕文字包含 '登入' 或對應的 class)
    # 註: 需要您按 F12 確認該按鈕的精確 Selector，以下為通用寫法嘗試
    try:
        login_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "header[id='header'] li:nth-child(5) a:nth-child(1)")))
        login_btn.click()
        print("已點擊登入按鈕")
    except:
        print("找不到登入按鈕，請確認是否已經在登入畫面，或手動點擊")

    # 3. 等待登入 Modal (彈窗) 出現，並輸入帳號
    print("等待登入框出現...")
    # 改用更穩定的 ID 或 Name 定位 (假設 ID 是 email 和 password)
    # 如果您確定 placeholder 是 '輸入信箱'，可用 CSS_SELECTOR，但要加 wait
    email_input = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "header[id='header'] li:nth-child(5) a:nth-child(1)")))
    email_input.clear()
    email_input.send_keys(ACCOUNT)
    print("已輸入帳號")

    # 4. 輸入密碼
    password_input = driver.find_element(By.CSS_SELECTOR, "input[placeholder='輸入密碼']")
    password_input.clear()
    password_input.send_keys(PASSWORD)
    print("已輸入密碼")

    button = driver.find_elements(By.CSS_SELECTOR, "input[value='登入']")
    button.click()

    # ==========================================
    # 關鍵：停下來讓人類介入
    # ==========================================
    print("\n" + "="*50)
    print("【請注意】程式已幫您填寫帳密。")
    print("請您現在手動輸入「圖形驗證碼」並按下「登入」按鈕。")
    print("登入成功後，請回來看這裡，程式將在 60 秒後或偵測到登入成功後繼續...")
    print("="*50 + "\n")

    # 5. 等待登入成功 (偵測 URL 變化或 Cookie 變化)
    # 這裡我們簡單等待您手動操作，直到頁面狀態改變
    WebDriverWait(driver, 120).until(
        # 假設登入成功後，登入按鈕會消失，或是出現「登出」字樣
        # 這裡簡單判斷：是否有拿到 session cookie
        lambda d: any(cookie['name'] == 'JSESSIONID' or 'TS' in cookie['name'] for cookie in d.get_cookies())
    )
    
    print("偵測到可能的登入成功！準備提取 Cookies...")
    time.sleep(3) # 再多等一下確保 Cookie 寫入

    # (接下來接續您的 Cookie 提取程式碼...)

except Exception as e:
    print(f"發生錯誤: {e}")
    # 發生錯誤時不要馬上關閉，讓您有機會看清楚發生什麼事
    input("按 Enter 鍵關閉瀏覽器...")
```

```python
import requests
import time
import random
from datetime import datetime, timedelta

# Selenium 相關模組
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ==========================================
# 第一階段：使用 Selenium 進行登入 (獲取 Cookies)
# ==========================================

print("正在啟動瀏覽器...")
# 初始化 Chrome 瀏覽器
driver = webdriver.Chrome()

try:
    # 1. 前往 CODIS 網站 (通常前往首頁或測站查詢頁面)
    target_url = "https://codis.cwa.gov.tw/Station"
    driver.get(target_url)
    
    print("---")
    print("【重要步驟】瀏覽器已開啟。")
    print("請您在瀏覽器視窗中，手動點擊「登入」，輸入帳號密碼並完成登入。")
    print("程式將等待 60 秒，或偵測到您登入完成後自動繼續...")
    print("---")

    # 2. 等待登入完成
    # 這裡設定一個條件：等待網址中不再包含 'Login' 字眼，或是等待特定的登入後元素出現
    # 為了彈性，這裡我們先用 "等待 URL 穩定" 或 "單純等待時間" 讓您手動操作
    # 實務上：您可以手動登入，只要在 60秒內完成即可
    time.sleep(10) # 給予緩衝時間讓頁面載入
    
    # 這裡使用 WebDriverWait 等待使用者登入成功
    # (判斷標準：假設登入後 URL 會變，或者您可以觀察 Cookie 的變化)
    # 這裡為了簡單起見，我們讓程式等待您手動操作，直到您回到 /Station 頁面或類似狀態
    WebDriverWait(driver, 60).until(
        lambda d: "_ga" in [c['name'] for c in d.get_cookies()] # 簡單檢查是否有 Cookie 生成
    )
    
    # 再次強制等待幾秒，確保所有 Session Cookies (如 TSxxxx) 都已寫入
    time.sleep(5)
    print("偵測到登入動作，正在提取 Cookies...")

    # 3. 提取 Selenium 的 Cookies
    selenium_cookies = driver.get_cookies()
    
    # 4. 提取 User-Agent (為了讓 Requests 偽裝得跟瀏覽器一模一樣)
    user_agent = driver.execute_script("return navigator.userAgent;")

finally:
    # 關閉瀏覽器 (Cookies 已經存在變數裡了，不需要瀏覽器了)
    driver.quit()
    print("瀏覽器已關閉。")

# ==========================================
# 第二階段：轉移權限給 Requests (建立 Session)
# ==========================================

# 1. 建立 Session
session = requests.Session()

# 2. 設定 Headers
# 使用剛才瀏覽器的 User-Agent，並補上 Referer
headers = {
    "User-Agent": user_agent,
    "Referer": "https://codis.cwa.gov.tw/station",
    "Origin": "https://codis.cwa.gov.tw",
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "X-Requested-With": "XMLHttpRequest"
}
session.headers.update(headers)

# 3. 將 Selenium 的 Cookies 注入到 Requests Session 中
for cookie in selenium_cookies:
    session.cookies.set(cookie['name'], cookie['value'], domain=cookie['domain'])

print(f"成功轉移 {len(selenium_cookies)} 個 Cookies 到 Session。")
print("開始執行高速爬取...")

# ==========================================
# 第三階段：原本的 Requests 爬蟲迴圈
# ==========================================

api_url = "https://codis.cwa.gov.tw/api/large_download_station"
start_date = datetime(2025, 1, 2)

for i in range(5):
    current_day = start_date + timedelta(days=i)
    
    # payload 設定
    payload = {
        "type": "hour",
        "stn_ID": "C0I480",
        "stn_type": "auto_C0",
        "start": current_day.strftime("%Y-%m-%dT00:00:00"),
        "end": current_day.strftime("%Y-%m-%dT23:59:59")
    }
    
    try:
        # 使用擁有登入權限的 session 發送 POST
        response = session.post(api_url, data=payload)
        
        # 檢查 WAF 攔截
        if "<title>Request Rejected</title>" in response.text:
            print(f"Data for {current_day.date()}: 被防火牆攔截 (Request Rejected)")
            # 如果這裡失敗，通常代表 User-Agent 不一致或 Cookie 缺漏
            break 
            
        if response.status_code == 200:
            try:
                data = response.json()
                # 檢查資料內容是否為空 (有些登入失效會回傳 200 但 data 是空的)
                if 'data' in data:
                    count = len(data['data'])
                    print(f"Data for {current_day.date()}: 成功取得 (資料筆數: {count})")
                else:
                    print(f"Data for {current_day.date()}: 請求成功但無資料 (可能是權限問題或當日無數據) - {data}")
            except Exception as e:
                print(f"Data for {current_day.date()}: JSON 解析失敗 (可能未登入成功) - {response.text[:50]}...")
        else:
            print(f"Data for {current_day.date()}: HTTP 狀態碼錯誤 {response.status_code}")

    except Exception as e:
        print(f"請求發生錯誤: {e}")

    # 隨機延遲
    time.sleep(random.uniform(2, 5))
```
