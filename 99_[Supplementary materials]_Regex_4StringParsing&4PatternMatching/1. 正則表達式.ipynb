正則表達式 - re.match 使用說明

```python
# match
'''
說明
re.match 會從字串的「開頭」開始比對，
比對不到，則回傳 None
變數名稱自己取
PHP, Javacript : 斜線是有意義的
re.match()第一個參數要放正規表達式，第二個放要剖析的字串
當re.match成功時，回傳的不是簡單的True，而是資訊豐富的Match物件
'''
regex01 = r'2[0-9]{3}\/[0-1]?[0-9]{1}\/([0-3]?[0-9])'  # 日期格式的正則表達式
string01 = "2025/10/08"  # 待匹配的字串
match01 = re.match(regex01, string01)  # 執行匹配
print(match01)  # 顯示匹配結果物件
print(match01[0])  # 顯示完整匹配內容
print(match01[1])  # 顯示第一個捕獲組內容

'''
補充:
match.group() 或 match.group(0) 是regex所代表的整個完整比對的字串，
match.group(1)是第一組()中的內容，
match.group(2)是第二組()中的內容...
'''
print(match01.group(0))  # 等同於 match01[0]
print(match01.group(1))  # 等同於 match01[1]
```

正則表達式 - re.findall 使用說明

```python
# findall
'''
說明
Re.findall 這個函式會掃描整個String，找出所有與Pattern相匹配的
re.findall 會將所有配對到的字串
回傳成一個 list
'''
regex02 = r'[0-9]+'  # 匹配一個或多個數字的正則表達式
string02 = "0944444444, 0955555555, 0966666666"  # 待匹配的字串，包含三個電話號碼
listMatch02 = re.findall(regex02, string02)  # 執行全局匹配
print(listMatch02)  # 顯示所有匹配結果的列表
print(listMatch02[0])  # 顯示第一個匹配結果
print(listMatch02[2])  # 顯示第三個匹配結果
```

正則表達式 - re.finditer 使用說明

```python
# finditer
'''
說明
re.finditer 會將所有配對到的字串
以迭代的方式呈現，逐一產出每一次成功匹配的Match物件，若沒有配對到，則回傳 None
Match這個變數是動態變化的，[0]操作對象在每一次迴圈中都是全新物件
'''
regex03 = r'[0-9]+'  # 匹配一個或多個數字的正則表達式
string03 = "0944444444, 0955555555, 0966666666"  # 待匹配的字串，包含三個電話號碼
iterableMatch03 = re.finditer(regex03, string03)  # 執行匹配，返回迭代器

# 遍歷迭代器，獲取每個匹配結果
for match in iterableMatch03:
    print(match[0])  # 顯示每個匹配的完整內容

# 注意：此時迭代器已經用完，再次打印會顯示迭代器物件本身
print(iterableMatch03)
```

正則表達式 - re.search 使用說明

```python
# search
'''
說明
re.search 會將整個字串進行搜尋，
但只會比對到第一組，也就是單一匹配: 一旦找到第一個匹配，函式就會停止搜尋，回傳結果。比對不到，則回傳 None
'''
regex04 = r'[a-zA-Z]([12])\d{8}'  # 匹配台灣身分證號碼格式的正則表達式
string04 = "A123456789, S299888777"  # 待匹配的字串，包含兩個身分證號碼

match04 = re.search(regex04, string04)  # 執行搜尋，返回第一個匹配結果
print(match04)  # 顯示匹配物件的詳細資訊
print(match04[0])  # 顯示完整匹配的字串
print(match04[1])  # 顯示第一個捕獲組的內容
```

正則表達式 - re.split 使用說明

```python
import re  # 確保導入re模組

# split
'''
說明
re.split 類似 string.split('separator')，
只是用正規表達式來作為 separator，
並回傳 list
'''
regex05 = r'\d'  # 匹配任何單個數字的正則表達式
string05 = "One1Two2Three3Four4"  # 待分割的字串，單詞間以數字分隔

listMatch05 = re.split(regex05, string05)  # 使用數字作為分隔符拆分字串
print(listMatch05)  # 顯示分割後的結果列表
```

正則表達式 - re.sub 使用說明

```python
# sub
'''
說明
re.sub(regex, replace_string, test_string)
將 regex 所代表的文字，改成 replace_string，文字來源是 test_string
\D: 任何非數字
'''
regex06 = r"\D"  # 匹配任何非數字字符的正則表達式
string06 = "5-20 #1314"  # 待處理的字串，包含數字和非數字字符

strResult = re.sub(regex06, "", string06)  # 將所有非數字字符替換為空字串
print(strResult)  # 顯示替換後的結果
```

 正則表達式中的環視語法

| 名稱 | 語法 | 說明 |
|------|------|------|
| 正向環視 | (?=) | 這位置右邊要出現什麼 |
| 正向環視否定 | (?!) | 這位置右邊不能出現什麼 |
| 反向環視 | (?<=) | 這位置左邊要出現什麼 |
| 反向環視否定 | (?<!) | 這位置左邊不能出現什麼 |


```python

# 環視 (例如去除中文字旁邊的空白)
'''
說明
\\s : 匹配任何空白字符(空格、制表符、換行符等)
(?![a-zA-Z]) : 正向環視否定，確保右側不是英文字母
'''

regex07 = r"\s(?![a-zA-Z])"  # 匹配後面不是英文字母的空白
string07 = "一 天 一 蘋 果 醫 生 遠 離 我。An apple a day keeps the doctor away."
strResult = re.sub(regex07, '', string07)
print(strResult)  # 輸出結果
```

正則表達式 - 環視應用於千分位格式化

```python
# 環視 (加入千分位)
'''
說明：
(?<=\d) : 反向環視，確保左側是數字
(?=(\d{3})+\b) : 正向環視，確保右側是3的倍數個數字直到單詞邊界
'''

regex08 = r'(?<=\d)(?=(\d{3})+\b)'  # 匹配需要插入千分位逗號的位置
string08 = '1234567890'
strResult = re.sub(regex08, ',', string08)
print(strResult)  # 輸出結果：1,234,567,890
```
