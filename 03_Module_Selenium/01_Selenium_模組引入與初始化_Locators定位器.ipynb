````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

# 設定 Options (瀏覽器層級設定)
options = Options()
options.add_argument("--start-maximized") #最大化視窗
options.add_argument("--disable-blink-features=AutomationControlled") # 避免被視為機器人
# 設定 User-Agent
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36") 
# options.add_argument("--headless") #無頭模式 (可選)

# 實例化 WebDriver
driver = webdriver.Chrome(options = options)

try:
	driver.get("https://www.ispan.com.tw/")
	print("Page Title:", driver.title)
finally:
	driver.quit()
````

### 透過CSS 選擇器定位元素

````python
# 匯入 webdriver 模組
from selenium import webdriver 

# 匯入 By 類別，用於定位元素
from selenium.webdriver.common.by import By 

# 匯入 sleep 函數，用於暫停執行
from time import sleep 

# 啟動 Chrome 瀏覽器
driver = webdriver.Chrome() 

# 設定目標網址
url = "https://www.ispan.com.tw/" 

# 前往指定網址
driver.get(url) 
# 驗證網頁標題是否包含特定字串
assert 'iSpan資展國際｜數位人才培育｜課程多元' in driver.title 

# 透過 CSS 選擇器尋找特定連結元素
elem = driver.find_element(By.CSS_SELECTOR, 'a[href="/longterm/"]')

# 點擊該元素
elem.click() 

# 驗證跳轉後的網頁標題
assert '就業養成 | iSpan資展國際' in driver.title 

# 暫停 5 秒
sleep(5) 

# 關閉瀏覽器視窗並結束驅動程序
driver.quit()
````

### 2. `By.PARTIAL_LINK_TEXT` (模糊比對)
比對超連結的部分文字。
```python
element = driver.find_element(By.PARTIAL_LINK_TEXT, "部分文字")
```

````python
# 匯入 selenium 的 webdriver 模組
from selenium import webdriver

# 匯入 By 模組，用於定位元素
from selenium.webdriver.common.by import By

# 匯入 sleep 函式，用於暫停執行
from time import sleep

# 啟動 Chrome 瀏覽器
driver = webdriver.Chrome()

# 設定目標網址
url = "https://www.ispan.com.tw/"

# 開啟網頁
driver.get(url)

# 驗證網頁標題是否包含特定文字
assert 'iSpan資展國際｜數位人才培育｜課程多元' in driver.title

#<a class="text_item__mb__cpycp" style="display:inline-block" href="/privacypolicy/">服務條款及隱私權保護政策</a>
elem = driver.find_element(By.PARTIAL_LINK_TEXT, '隱私權')

# 點擊該元素
elem.click()

# 暫停 5 秒
sleep(5)

# 關閉瀏覽器並結束視窗
driver.quit()
````
