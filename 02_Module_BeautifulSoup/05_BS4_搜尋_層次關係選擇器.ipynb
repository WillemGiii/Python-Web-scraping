```python
HTML=r'''<!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Explain child Selector</title>
        </head>
        <body>
            <div class="card">
            <!-- 第一層 (兒子) -->
            <p>我是摘要 (Direct Child)</p>
            
            <div class="content">
                <!-- 第二層 (孫子) -->
                <p>我是內文細節 (Grandchild)</p>
            </div>
        </div>
        </body>
        </html>'''


```

```python
from bs4 import BeautifulSoup as bs4
soup = bs4(HTML, 'lxml')
```

```python
# 找出 .card 裡面的「所有」 p
soup.select("div.card p")
```

```python
# 找出 .card 「第一層」的 p
soup.select("div.card > p")
```
