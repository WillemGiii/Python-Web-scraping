```python
import requests
import os
from dotenv import load_dotenv

# 1. 環境變數設定
load_dotenv()
AUTHORIZATION = os.getenv("AUTHORIZATION")

# 2. 定義參數
# 注意：dataid 是路徑的一部分，不放入 params
dataid = 'O-A0001-001' 

# 3. 準備 Query Parameters (查詢參數)
# 這些參數會被 requests 自動轉換為 url?key=value 的形式
params = {
    'Authorization': AUTHORIZATION,
    'StationId': 'C0D550',  # 指定測站
    'format': 'JSON',      # 指定回傳格式 (注意大小寫，API文件通常建議大寫，雖小寫常通融)
    'timeFrom': '2023-04-01T00:00:00',
    'timeTo' : '2023-04-05T00:00:00'
}


# 4. 組合 URL 並發送請求
# 修正 Base URL 為 /api/v1/rest/datastore
base_url = "https://opendata.cwa.gov.tw/api/v1/rest/datastore"
url = f"{base_url}/{dataid}"  # 使用 f-string 將 dataid 拼接到路徑中

# 發送 GET 請求
res = requests.get(url, params=params)

# --- 觀察回應 ---

print(f"實際請求的 URL: {res.url}") 
# 教學重點：讓學生看這行，確認 URL 是否為 .../C-B0024-001?Authorization=...

print(f"Status Code: {res.status_code}")

if res.status_code == 200:
    data = res.json()
    # 稍微印出一點資料驗證
    pprint.pprint(data)

else:
    print("請求失敗")
    print(res.text)
```
