Interact with elements by Hover 透過滑鼠懸停來互動

```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

# 1. 設定 HTML 檔案路徑
file_path = "file://" + os.path.abspath("action_chains_demo.html")

driver = webdriver.Chrome()

driver.get(file_path)

button = driver.find_element(By.ID, "cookie-btn")
button.click()

sleep(3)

# 定位父選單與子選單
menu = driver.find_element(By.CSS_SELECTOR, '#menu-hover')
# 注意 : 此時子選單 #submenu 是不可見的
submenu = driver.find_element(By.CSS_SELECTOR, '#link-option-1')

# 實例化 ActionChains
actions = ActionChains(driver)

# 建立一串動作(Queue)
# 1. 移動到父選單 (觸發 Hover 效果)
actions.move_to_element(menu)
# 2. 點擊子選單
actions.click(submenu)
# 3. 執行動作
actions.perform()
# 觀念 : 在呼叫 .perform() 之前，瀏覽器甚麼都不會做
actions.perform()
print("已執行懸停並點擊子選單")
time.sleep(5)

driver.quit()
```

Interact with elements by dragging 透過拖曳來互動

```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

# 1. 設定 HTML 檔案路徑
file_path = "file://" + os.path.abspath("action_chains_demo.html")

driver = webdriver.Chrome()

driver.get(file_path)
sleep(3)
button = driver.find_element(By.ID, "cookie-btn")
button.click()
sleep(3)

source_ele = driver.find_element(By.ID, "draggable")
target_ele = driver.find_element(By.ID, 'droppable')

#重新實例化 ActionChains (或清除之前的 Actions)
drag_actions = ActionChains(driver)

# 使用內建的 drag_and_drop 方法
# 這是一個便利方法 ， 等同於 click_and_hold -> move_to -> release
drag_actions.drag_and_drop(source_ele, target_ele)
drag_actions.perform()
print("拖曳動作已完成")

#讓大家看一下最終成果
time.sleep(5)

driver.quit()
```
